````python
import numpy as np
import torch
import torch_geometric
import graph_tool.all as gt
import graft
import warnings
warnings.filterwarnings("ignore")
````

# 3. 기본사용

````python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
g = torch_geometric.data.Data(
    edge_index = links
)
````

undirected graf임

````python
graft.plot(g)
````

노드이름을 정의함

````python
graft.plot(
    g,
    node_names=[0,1,2,3,4]
)
````

````python
graft.plot(
    g,
    node_names=["A","B","C","D","E"],
)
````

이름을 쓰기 싫다면 공백을 넣을수도있음.

````python
graft.plot(
    g,
    node_names=["","","","A",""]
)
````

````python
graft.plot(
    g,
    node_names=["A","B","C","D","E"],
    node_sizes=[1,1,2,2,3],
)
````

````python
graft.plot(
    g,
    node_names=["A","B","C","D","E"],
    node_colors=[0,0,1,2,2],
)
````

````python
# graft.plot(
#     g,
#     node_names=["A","B","C","D","E"],
#     node_colors=[1,1,1,2,2], # 버그.. 카테고리일 경우 node_colors는 0부터 시작해야함
# )
````

````python
graft.plot(
    g,
    node_names=["A","B","C","D","E"],
    node_colors=np.array([0,0,1,2,2])/2,
)
````

````python
graft.plot(
    g,
    node_names=["A","B","C","D","E"],
    node_colors=np.array([0,0,1,2,2])/2,
    node_sizes=np.array([0,0,1,2,2])/2,
)
````

# 4. 옵션

````python
#graft.plot(g)
gt_graph, draw_options = graft.setup_graph_draw(g)
gt.graph_draw(gt_graph, **draw_options)
````

````python
# graft.plot(
#     g,
#     node_names=[0,1,2,3,4]
# )
gt_graph, draw_options = graft.setup_graph_draw(g)
gt_graph, draw_options = graft.map_node_names(gt_graph,draw_options,node_names=[0,1,2,3,4])
gt.graph_draw(gt_graph, **draw_options)
````

````python
# graft.plot(
#     g,
#     node_names=["A","B","C","D","E"],
# )
gt_graph, draw_options = graft.setup_graph_draw(g)
gt_graph, draw_options = graft.map_node_names(gt_graph,draw_options,node_names=["A","B","C","D","E"])
gt.graph_draw(gt_graph, **draw_options)
````

````python
# graft.plot(
#     g,
#     node_names=["","","","A",""]
# )
gt_graph, draw_options = graft.setup_graph_draw(g)
gt_graph, draw_options = graft.map_node_names(gt_graph,draw_options,node_names=["","","","A",""])
gt.graph_draw(gt_graph, **draw_options)
````

````python
# graft.plot(
#     g,
#     node_names=["A","B","C","D","E"],
#     node_sizes=[1,1,2,2,3],
# )
gt_graph, draw_options = graft.setup_graph_draw(g)
gt_graph, draw_options = graft.map_node_names(gt_graph,draw_options,node_names=["","","","A",""])
gt_graph, draw_options = graft.map_node_sizes(gt_graph,draw_options,node_sizes=[1,1,2,2,3])
gt.graph_draw(gt_graph, **draw_options)
````

````python
# graft.plot(
#     g,
#     node_names=["A","B","C","D","E"],
#     node_colors=[0,0,1,2,2],
# )
gt_graph, draw_options = graft.setup_graph_draw(g)
gt_graph, draw_options = graft.map_node_names(gt_graph,draw_options,node_names=["A","B","C","D","E"])
gt_graph, draw_options = graft.map_node_colors(gt_graph,draw_options,node_colors=[0,0,1,2,2])
gt.graph_draw(gt_graph, **draw_options)
````

````python
# graft.plot(
#     g,
#     node_names=["A","B","C","D","E"],
#     node_colors=np.array([0,0,1,2,2])/2,
# )
gt_graph, draw_options = graft.setup_graph_draw(g)
gt_graph, draw_options = graft.map_node_names(gt_graph,draw_options,node_names=["A","B","C","D","E"])
gt_graph, draw_options = graft.map_node_colors(gt_graph,draw_options,node_colors=np.array([0,0,1,2,2])/2)
gt.graph_draw(gt_graph, **draw_options)
````

````python
# graft.plot(
#     g,
#     node_names=["A","B","C","D","E"],
#     node_colors=np.array([0,0,1,2,2])/2,
#     node_sizes=np.array([0,0,1,2,2])/2,
# )
gt_graph, draw_options = graft.setup_graph_draw(g)
gt_graph, draw_options = graft.map_node_names(gt_graph,draw_options,node_names=["A","B","C","D","E"])
gt_graph, draw_options = graft.map_node_colors(gt_graph,draw_options,node_colors=np.array([0,0,1,2,2])/2)
gt_graph, draw_options = graft.map_node_sizes(gt_graph,draw_options,node_sizes=np.array([0,0,1,2,2])/2)
gt.graph_draw(gt_graph, **draw_options)
````
