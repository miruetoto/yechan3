# 1. 비트시그널

$$y_t = \cos(62\pi t) + \cos(58\pi t)$$

```r
t = 1:1000/2000
y1 = cos(62*pi*t) 
y2 = cos(58*pi*t)
y = y1 + y2 
fig1 = gglite() + line(t,y1,col=2) + line(t,y,alpha=0) + ylab("") + xlab("")
fig2 = gglite() + line(t,y2,col=4) + line(t,y,alpha=0) + ylab("") + xlab("")
fig3 = gglite() + line(t,y1+y2,) + line(t,y1,col=2,alpha=0.5) + line(t,y2,col=4,alpha=0.5) + ylab("") + xlab("")
```

```r
fig = fig1/fig2/fig3
fig
```

```r
ggsave("beat_signal.pdf",fig)
```
