# 

**About this blog**

This blog was created for my personal research, study and lecture
preparation. Therefore, the contents of the blog can be thought of as my
practice notes. As a result, sometimes the content of a post may be left
unstructured or unfinished. The blog is named after my favorite essay
‘신록예찬’, which is also a nickname I use informally. You can check the
written article in the sidebar on the left. It is a great honor for me
if these posts can help others to learn and research.
