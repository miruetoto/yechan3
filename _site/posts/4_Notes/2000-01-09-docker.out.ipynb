# **\[Note\]** docker

`-` 우분투에서 설치

ref: <https://docs.docker.com/engine/install/ubuntu/>

`-` 기본명령들

``` bash
docker run penguin
docker stop penguin
docker image pull penguin 
docker container start penguin
docker container run penguin
```

`-` Docker에서 다운로드한 이미지와 실행 중인 컨테이너를 확인하기 위한
명령어를 소개하겠습니다:

1.  **이미지 관련 명령어**:

    -   **docker images**: 현재 시스템에 다운로드된 모든 도커 이미지를
        확인할 수 있습니다.

    -   **docker image ls**: “docker images”와 동일한 결과를 출력합니다.

    -   **docker image ls <이미지_이름>**: 특정 이미지의 버전을 명시하여
        해당 이미지의 정보를 확인할 수 있습니다.

    -   **docker image history <이미지_이름>**: 특정 이미지의 레이어별
        변경 내역을 확인할 수 있습니다.

2.  **컨테이너 관련 명령어**:

    -   **docker ps**: 현재 실행 중인 컨테이너의 목록을 확인합니다.

    -   **docker ps -a**: 모든 컨테이너의 목록을 확인합니다 (실행 중인
        것과 중지된 것 모두 포함).

    -   **docker container ls**: “docker ps”와 동일한 결과를 출력합니다.

    -   **docker container ls -a**: “docker ps -a”와 동일한 결과를
        출력합니다.

    -   **docker container inspect <컨테이너_ID 또는 이름>**: 특정
        컨테이너의 자세한 정보를 확인할 수 있습니다.

    -   **docker logs <컨테이너_ID 또는 이름>**: 컨테이너의 로그를
        확인할 수 있습니다.

    -   **docker container stats**: 실행 중인 컨테이너의 실시간 리소스
        사용량을 모니터링할 수 있습니다.

이 명령어들을 사용하여 이미지와 컨테이너에 대한 정보를 확인하고 관리할
수 있습니다.

`-` `docker run -ti --rm r-base` 명령어에서 사용된 옵션들은 다음과 같은
의미를 갖습니다:

1.  **-ti**: 이 옵션은 두 개의 플래그인 `-t`와 `-i`를 결합한 것입니다.
    -   `-t` (또는 `--tty`): 이 플래그는 유사 터미널을 할당하고 터미널
        상호작용을 가능하게 합니다. 컨테이너 내부에서 터미널 환경을
        사용할 수 있게 해줍니다.
    -   `-i` (또는 `--interactive`): 이 플래그는 컨테이너와 상호작용할
        수 있도록 표준 입력(STDIN)을 열어줍니다. 컨테이너 내부에서
        입력을 주고 받을 수 있게 해줍니다.
2.  **–rm**: 이 옵션은 컨테이너가 실행을 마칠 때 자동으로 컨테이너를
    삭제합니다. 컨테이너가 종료되면 그 내용과 상태는 모두 삭제됩니다.
    주로 임시 작업을 수행하는데 사용됩니다.

따라서 `docker run -ti --rm r-base` 명령어는 R 언어 기반의 컨테이너를
실행하는 명령어로, 터미널 환경을 제공하며 컨테이너가 종료될 때
컨테이너를 자동으로 삭제합니다. 이를 통해 R 언어를 실행하고 테스트하거나
실험할 수 있는 환경을 구성할 수 있습니다.

`-`
`docker run -ti --rm -v "$PWD":/home/docker -w /home/docker -u docker r-base R CMD check`
명령어에서 사용된 각 옵션에 대한 설명은 다음과 같습니다:

1.  **-ti**: 이 옵션은 터미널 상호작용을 위한 옵션입니다. `-t`는 유사
    터미널 할당을 의미하고, `-i`는 터미널 입력 상호작용을 허용하는 것을
    의미합니다.

2.  **–rm**: 컨테이너가 실행을 마칠 때 자동으로 컨테이너를 삭제합니다.

3.  **-v “\$PWD”:/home/docker**: 이 옵션은 호스트 시스템의 현재 작업
    디렉토리를 컨테이너 내의 `/home/docker` 디렉토리와 공유
    마운트합니다. 이렇게 함으로써 호스트 시스템의 파일을 컨테이너
    내부에서 사용할 수 있게 됩니다.

4.  **-w /home/docker**: 컨테이너가 실행될 작업 디렉토리를
    `/home/docker`로 설정합니다. 즉, 컨테이너가 실행될 때 기본
    디렉토리가 `/home/docker`가 됩니다.

5.  **-u docker**: 이 옵션은 컨테이너 안에서 사용자를 변경하여 실행하는
    것을 의미합니다. `docker`라는 사용자로 컨테이너를 실행하게 됩니다.

6.  **r-base**: Docker 이미지 이름입니다. 여기서는 R 언어를 기반으로한
    이미지를 사용합니다.

7.  **R CMD check**: 실제로 실행되는 컨테이너 내부의 명령어입니다. 이
    경우 R 패키지를 체크하는 R 명령어 `R CMD check`가 실행됩니다.

이 명령어는 호스트 시스템의 현재 작업 디렉토리에 있는 R 패키지를
컨테이너 내부에서 `R CMD check` 명령으로 체크하고자 할 때 사용될 수
있습니다.

`-` “docker container start penguin”과 “docker container run penguin” 두
명령어의 차이점은 다음과 같습니다:

1.  **docker container start penguin**:
    -   이 명령은 이미 생성된 (stopped 또는 exited 상태의) 컨테이너를
        시작시킬 때 사용됩니다.
    -   컨테이너의 상태를 변경하지 않고, 중단된 상태에서 다시
        실행합니다.
    -   이미지에서 컨테이너를 생성한 후, 실행 중인 컨테이너를 중단하고
        다시 시작할 때 유용합니다.
2.  **docker container run penguin**:
    -   이 명령은 도커 이미지를 기반으로 새로운 컨테이너를 생성하고
        실행시킬 때 사용됩니다.
    -   이미지에서 새로운 컨테이너를 생성하고, 컨테이너를 시작합니다.
    -   이미 실행 중인 컨테이너가 아닌, 새로운 컨테이너를 생성하고
        실행할 때 사용됩니다.

요약하면, “docker container start penguin”은 이미 생성된 컨테이너를 다시
시작하고, “docker container run penguin”은 새로운 컨테이너를 생성하고
시작하는 데 사용됩니다.
