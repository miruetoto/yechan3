```python
titles = [f.split('/')[-1] for f in fnames]
titles
```

```python
fig,axs = plt.subplots(len(fnames))
fig.set_size_inches(10,20)
for i in range(len(fnames)):
    signal = pd.read_csv(fnames[i],header=None)
    axs[i].plot(signal[:5000])
    axs[i].set_title(titles[i])
```
