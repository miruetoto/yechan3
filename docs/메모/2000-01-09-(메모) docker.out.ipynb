# 1. 기본명령어

`-` 기본명령들

``` bash
## 이미지 
docker images 
docker image pull penguin 
docker run {이미지이름}
docker image ls {이미지이름} # 특정 이미지의 버전을 명시하여 해당 이미지의 정보를 확인.
docker image history {이미지이름} # 특정 이미지의 레이어별 변경 내역을 확인.

## 컨테이너
docker ps 
docker ps -a 
docker start {컨테이너이름or컨테이너ID}
docker stop {컨테이너이름or컨테이너ID}
docker rm {컨테이너이름or컨테이너ID} # 컨테이너가 중단된 상태에서 실행해야함
docker container inspect {컨테이너이름or컨테이너ID} # 특정 컨테이너의 자세한 정보를 확인.
docker logs {컨테이너이름or컨테이너ID} # 컨테이너의 로그를 확인.
docker container stats # 실행 중인 컨테이너의 실시간 리소스 사용량을 모니터링.
```

# Appendix

`-` 기본적으로는 `docker run {옵션들} {이미지이름} {내부명령}`와 같은
포맷임

`-` `docker run -ti --rm r-base`

1.  **-ti**: 이 옵션은 두 개의 플래그인 `-t`와 `-i`를 결합. –\> 터미널
    쓸려면 필수임.
    -   `-t` (또는 `--tty`): 이 플래그는 유사 터미널을 할당하고 터미널
        상호작용을 가능하게 함. 컨테이너 내부에서 터미널 환경을 사용할
        수 있게 해줌.
    -   `-i` (또는 `--interactive`): 이 플래그는 컨테이너와 상호작용할
        수 있도록 표준 입력(STDIN)을 열어줌. 컨테이너 내부에서 입력을
        주고 받을 수 있게 해줌.
2.  **–rm**: 이 옵션은 컨테이너가 실행을 마칠 때 자동으로 컨테이너를
    삭제.

`-`
`docker run -ti --rm -v "$PWD":/home/docker -w /home/docker -u docker r-base R CMD check`

1.  **-ti**: 이 옵션은 터미널 상호작용을 위한 옵션

2.  **–rm**: 컨테이너가 실행을 마칠 때 자동으로 컨테이너를 삭제.

3.  **-v “\$PWD”:/home/docker**: 이 옵션은 호스트 시스템의 현재 작업
    디렉토리를 컨테이너 내의 `/home/docker` 디렉토리와 공유 마운트.
    이렇게 함으로써 호스트 시스템의 파일을 컨테이너 내부에서 사용할 수
    있게 됨.

4.  **-w /home/docker**: 컨테이너가 실행될 작업 디렉토리를
    `/home/docker`로 설정. 즉, 컨테이너가 실행될 때 기본 디렉토리가
    `/home/docker`가 된다.

5.  **-u docker**: 이 옵션은 컨테이너 안에서 사용자를 변경하여 실행하는
    것을 의미. `docker`라는 사용자로 컨테이너를 실행.

6.  **r-base**: Docker 이미지 이름.

7.  **R CMD check**: 실제로 실행되는 컨테이너 내부의 명령어. 이 경우 R
    패키지를 체크하는 R 명령어 `R CMD check`가 실행.

`-` `run` vs `start`

-   run: 이미지 -\> 컨테이너
-   start: 중단된 컨테이너 재시작
