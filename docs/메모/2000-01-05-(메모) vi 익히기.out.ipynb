# 모드

## `esc`: normal 모드

`-` 기능1: 되돌리기, 되돌리기를 되돌리기

`-` 기능2: 커서이동 (기본) – 방향키, 쉬프트+방향키, home/end,
pageup/pagedown 으로 대체가능

-   `k`,`j`,`h`,`l`: 상,하,좌,우
-   `w`,`b`: 다음 단어의 시작으로 이동, 이전 단어의 시작으로 이동
-   `0` (숫자 0), `$`: 줄의 맨 앞, 맨 끝으로 이동
-   `control + d`, `control + u`: 반페이지씩 page down, page up
-   `{`, `}`: 문단 시작, 끝으로 이동

`-` 기눙3: 커서이동 (고급)

-   `%`: 현재 괄호의 짝으로 이동
-   `e`: 현재단어의 끝으로 이동
-   `gg`: 파일의 첫 줄로 이동
-   `G`: 파일의 마지막 줄로 이동

`-` 기능4: 다른모드로 진입하기 위한 중간허브

-   `v`,`V`: 선택모드, 줄단위선택모드로 변환
-   `i`, `I`, `a`, `A`: 커서앞, 줄맨앞, 커서뒤, 줄맨뒤에서 삽입모드로
    전환
-   `r`, `R`: 한글자 수정후 복귀, 수정모드로 진입

`-` 기능5: 들여쓰기, 내어쓰기

-   `> + Enter`, `< + Enter`: 들여쓰기, 내어쓰기

## `i`: 삽입모드

## `d`: 삭제모드

## `:`: 명령모드

## `/`: 검색모드

## `v`: 비주얼모드(선택모드)

## `R`: 수정모드
