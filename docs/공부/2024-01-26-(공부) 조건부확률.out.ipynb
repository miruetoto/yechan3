# (공부) 조건부확률, 조건부기대값

신록예찬  
2024-01-26

> $\mathbb{E}[Y|X]$의 의미를 알아보자.

# 1. 엄밀하지 않은 정의

`# 정의` – 조건부확률(이산형)

$$P(Y=y|X=x):=\frac{P(X=x,Y=y)}{P(X=x)}$$

`#`

`# 정의` – 조건부pdf(이산형,연속형)

$$f_{2|1}(y|x):=\frac{f_{1,2}(x,y)}{f_1(x)}$$

`#`

`-` 함수 $f_{2|1}(y|x)$ 와 $f_{2}(y)$는 모두 확률변수 $Y$의 pdf 조건을
만족한다는 것을 주의하자. 즉 아래가 성립한다.

-   $\int_{-\infty}^{\infty} f_{2}(y)dy=1$
-   $\int_{-\infty}^{\infty} f_{2|1}(y|x)dy=1$

`# 정의` – 조건부평균(연속형)

$$\mathbb{E}[Y|X=x]:=\int y f_{2|1}(y|x)dy $$

`#`

`# 정의할 수 없음` – $\mathbb{E}(Y|X)$

학부수준에서는 엄밀하게 정의할수는 없으며, 개념적으로 접근한다.

1.  $\mathbb{E}(Y|X=x)$는 $x$의 값에 따라서 그 값이 바뀌는 함수로 해석할
    수 있다.
2.  1을 다시 표현하면 $h(x):=\mathbb{E}(Y|X=x)$를 만족하는 적당한 $h$가
    있다는 의미.
3.  2번에서 정의된 $h$에 입력을 $x$가 아니라 $X$로 준다면 $h(X)$와 같은
    확률 변수를 생각할 수 있음. [1]
4.  3번과 같은 확률변수를 $\mathbb{E}(Y|X)$로 정의하자는 의미

학부수준에서는 이정도가 설명의 끝이며 $\mathbb{E}(Y|X)$를 수식으로
명확하게 정의하진 않는다.

`#`

`# 정의할 수 없음` – $\mathbb{V}(Y|X)$

$\mathbb{E}(Y|X)$와 비슷한 방식으로 피해서 정의한다.

`#`

`# 예제` – $\mathbb{E}[\mathbb{E}[Y|X]]=\mathbb{E}[Y]$ 의 증명 ($\star$)

(풀이)

헷갈리니까 아래와 같이 치환한다.

-   $\mathbb{E}[Y|X]=h(X)$
-   $\mathbb{E}[Y|X=x]=h(x)$

그러면 좌변은 아래와 같이 된다.

-   $\mathbb{E}[h(X)]=\int h(x) f_1(x)dx = \int \mathbb{E}[Y|X=x] f_1(x)dx$

그런데 $\mathbb{E}[Y|X=x]$는 조건부확률의 정의에 의하여
$\mathbb{E}[Y|X=x]=\int y f_{2|1}(y|x)dy$ 이다. 이걸 그대로 넣으면

-   $\int \mathbb{E}[Y|X=x] f_1(x)dx=\int \int y f_{2|1}(y|x)dy f_1(x)dx=\int \int y f_{2|1}(y|x)f_1(x)dydx$

이제 조건부 pdf의 정의를 쓰면 $f_{2|1}(y|x)f_1(x)=f_{1,2}(x,y)$가 된다.
따라서

-   $\int \int y f_{2|1}(y|x)f_1(x)dydx=\int\int y f_{1,2}(x,y)dydx$

푸비니의 정리를 쓰면

-   $\int\int y f_{1,2}(x,y)dydx=\int\int y f_{1,2}(x,y)dxdy$

$y$를 밖으로 빼고 marginal pdf의 정의[2]를 이용하면

-   $\int\int y f_{1,2}(x,y)dxdy=\int y\int f_{1,2}(x,y)dxdy=\int y f_2(y)dy = \mathbb{E}[y]$

그래서 증명이 되었음.

# 2. 엄밀한정의

`# 예제` - 주사위던지기

`-` 아래와 같은 상황을 가정하자.

-   $\Omega = \{1,2,3,\dots,6\}$
-   ${\cal F} = 2^\Omega$
-   $P: {\cal F} \to [0,1]$ such that $P(\{\omega\})=\frac{1}{6}$.

이제 $(\Omega, {\cal F})$ 위에서 아래와 같은 확률변수
$X: \Omega \to \mathbb{R}$을 정의하자.

-   $X(\{1\}) =1$
-   $X(\{2\}) =2$
-   $X(\{3\}) =3$
-   $X(\{4\}) =4$
-   $X(\{5\}) =5$
-   $X(\{6\}) =6$

여기까지는 익숙한 셋팅이며, 이럴때

-   $X$ is r.v. on $(\Omega, {\cal F})$
-   $X \in {\cal F}$

와 같이 표현한다.

`-` 이제 여기에서 아래와 같이 $\Omega$를 disjoin한 $\{B_i\}$로 나누자.

-   $B_1 = \{1,2\}$
-   $B_2 = \{3,4,5,6\}$

그리고 $B_1$, $B_2$를 measurable하게 만드는 새로운 시그마필드
${\cal F}^*=\sigma(B_1,B_2)$를 아래와 같이 정의하자.

-   ${\cal F}^* = \{\emptyset, B_1, B_2, \Omega\}$[3]

이제 아래와 같은 $Y: \Omega\to\mathbb{R}$를 고려하자.

-   $Y(\{1\}) = 1.5$
-   $Y(\{2\}) = 1.5$
-   $Y(\{3\}) = 4.5$
-   $Y(\{4\}) = 4.5$
-   $Y(\{5\}) = 4.5$
-   $Y(\{6\}) = 4.5$

여기에서 $Y$는 $(\Omega, {\cal F})$ 에서의 확률변수이기도 하지만
$(\Omega, {\cal F}^*)$에서의 확률변수이기도 하다. 즉

-   $Y$ is r.v. on $(\Omega, {\cal F})$
-   $Y \in {\cal F}$
-   $Y$ is r.v. on $(\Omega, {\cal F}^*)$
-   $Y \in {\cal F}^*$

이다.[4]

또 $Y$는 아래와 같은 특징이 있다.

-   $\forall A \in {\cal F}^*:~\int_A X dP = \int_A Y dP.$

이것을 조금만 따져보자. $\int_A X dP = \int_\Omega X \mathbb{1}_A dP$
임을 이용하면,

-   case1: $A = \emptyset \Rightarrow 0=0$:
-   case2:
    $A = B_1 \Rightarrow \frac{1+2+0+0+0+0}{6}=\frac{1.5+1.5+0+0+0+0}{6}$
-   case3:
    $A = B_2 \Rightarrow \frac{0+0+3+4+5+6}{6}=\frac{0+0+4.5+4.5+4.5+4.5}{6}$
-   case4:
    $A = \Omega \Rightarrow \frac{1+2+3+4+5+6}{6}=\frac{1.5+1.5+4.5+4.5+4.5+4.5}{6}$

이다.

`-` 이러한 조건을 만족하는 확률변수 $Y$를 conditional expectation of $X$
given ${\cal F}^*$이라고 한다.[5]

`#`

`# 정의` – (Klenke 2013, Def 8.11)
$X \in {\cal L}^1(\Omega, {\cal F}, P)$ 이라고 하자. [6] 그리고
${\cal F}^*$는

-   ${\cal F}^* \subset {\cal F}$

을 만족하는 $\Omega$에서의 시그마필드라고 하자. 확률변수 $Y$ 가 아래를
만족한다고 하자.

1.  $Y \in {\cal F}^*$
2.  $\forall A \in {\cal F}^*:~ \int_A X dP = \int_A Y dP$

그러면 $Y$는 conditional expectation of $X$ given ${\cal F}^*$라고 하고
기호로는

$$Y:=\mathbb{E}[X|{\cal F}^*]$$

로 표현한다.

`#`

`-` $Y$는 $(\Omega,{\cal F}^*)$에서의 확률변수이기도 하지만 일단
$Y \in (\Omega, {\cal F})$ 이기도 하다. 가측공간 $(\Omega,{\cal F}^*)$
에서는 적절한 확률측도를 정의한적이 없지만 $(\Omega,{\cal F})$ 에서는
확률측도 $P$가 이미 정의되어 있다. 이를 이용하여 아래를 계산해보자.

-   $\mathbb{E}[Y]=\int Y dP$

이것을 정리하면

-   $\frac{1.5+1.5+4.5+4.5+4.5+4.5}{6}$

가 되어서 $\mathbb{E}[X]$와 같다. 따라서

-   $\mathbb{E}[X] = \mathbb{E}[\mathbb{E}[X|{\cal F}^*]]$

이다. 여기에서 $\mathbb{E}[X|{\cal F}^*]$은 일단 그 자체로 하나의 기호라
생각하자.

`# Thm` – (Klenke 2013, Thm 8.12)
$X \in {\cal L}^1(\Omega, {\cal F}, P)$ 이라고 하자. [7] 그리고
${\cal F}^*$는

-   ${\cal F}^* \subset {\cal F}$

을 만족하는 $\Omega$에서의 시그마필드라고 하자. 아래를 만족하는 확률변수
$Y$는 ($P$에 대하여) 거의 유일하게 존재한다.

1.  $Y \in {\cal F}^*$
2.  $\forall A \in {\cal F}^*:~ \int_A X dP = \int_A Y dP$

`#`

`# 정의` – (Klenke 2013, Def 8.13) 아래와 같은 두 확률변수를 정의하자.

-   $X$ is rv on $(\Omega, {\cal F})$ with $\int X dP <\infty$.[8]
-   $Y$ is rv on $(\Omega, {\cal F})$.

아래를 정의하자.

$$\mathbb{E}[X|Y]:=\mathbb{E}[X|\sigma(Y)].$$
