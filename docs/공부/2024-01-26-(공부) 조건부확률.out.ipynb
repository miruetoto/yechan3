`# 예제` - 주사위던지기

`-` 아래와 같은 상황을 가정하자.

-   $\Omega = \{1,2,3,\dots,6\}$
-   ${\cal F} = 2^\Omega$
-   $P: {\cal F} \to [0,1]$ such that $P(\{\omega\})=\frac{1}{6}$.

이제 $(\Omega, {\cal F})$ 위에서 아래와 같은 확률변수
$X: \Omega \to \mathbb{R}$을 정의하자.

-   $X(\{1\}) =1$
-   $X(\{2\}) =2$
-   $X(\{3\}) =3$
-   $X(\{4\}) =4$
-   $X(\{5\}) =5$
-   $X(\{6\}) =6$

여기까지는 익숙한 셋팅이며, 이럴때

-   $X$ is r.v. on $(\Omega, {\cal F})$
-   $X \in {\cal F}$

와 같이 표현한다.

`-` 이제 여기에서 아래와 같이 $\Omega$를 disjoin한 $\{B_i\}$로 나누자.

-   $B_1 = \{1,2\}$
-   $B_2 = \{3,4,5,6\}$

그리고 $B_1$, $B_2$를 measurable하게 만드는 새로운 시그마필드
${\cal F}^*=\sigma(B_1,B_2)$를 아래와 같이 정의하자.

-   ${\cal F}^* = \{\emptyset, B_1, B_2, \Omega\}$[1]

이제 아래와 같은 $Y: \Omega\to\mathbb{R}$를 고려하자.

-   $Y(\{1\}) = 1.5$
-   $Y(\{2\}) = 1.5$
-   $Y(\{3\}) = 4.5$
-   $Y(\{4\}) = 4.5$
-   $Y(\{5\}) = 4.5$
-   $Y(\{6\}) = 4.5$

여기에서 $Y$는 $(\Omega, {\cal F})$ 에서의 확률변수이기도 하지만
$(\Omega, {\cal F}^*)$에서의 확률변수이기도 하다. 즉

-   $Y$ is r.v. on $(\Omega, {\cal F})$
-   $Y \in {\cal F}$
-   $Y$ is r.v. on $(\Omega, {\cal F}^*)$
-   $Y \in {\cal F}^*$

이다.[2]

또 $Y$는 아래와 같은 특징이 있다.

-   $\forall A \in {\cal F}:~\int_A X dP = \int_A Y dP.$[3]

이러한 조건을 만족하는 확률변수 $Y$를 conditional expectation of $X$
given ${\cal F}^*$이라고 한다.[4]

[2] 이렇게 보니까 ${\cal F}^* \subset {\cal F}$ 임을 감안하면
$Y \in {\cal F}^* \Rightarrow Y \in {\cal F}$ 임이 당연하게 느껴진다.
$Y \in {\cal F}$와 같은 기호표현이 너무 이상하게 느껴졌는데, 여기에서
의문이 조금 풀리는 것 같음.

[3] 좌변은 $\frac{1+2+3+4+5+6}{6}$ 이고 우변은
$\frac{1.5+1.5+4.5+4.5+4.5+4.5}{6}$
