````python
import numpy as np
import torch
import torch_geometric
import graph_tool.all as gt
import graft
import warnings
warnings.filterwarnings("ignore")
````

# 3. 기본사용

````python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
g = torch_geometric.data.Data(
    edge_index = links
)
````

undirected graf임

````python
graft.plot(g)
````

노드이름을 정의함

````python
graft.plot(
    g,
    node_names=[0,1,2,3,4]
)
````

이름을 쓰기 싫다면 공백을 넣을수도있음.

````python
graft.plot(
    g,
    node_names=["","","","A",""]
)
````

````python
graft.plot(
    g,
    node_names=["A","B","C","D","E"],
)
````

````python
graft.plot(
    g,
    node_names=["A","B","C","D","E"],
    node_sizes=[0,0,1,2,3],
    node_colors=[0,0,0,1,1]
)
````
