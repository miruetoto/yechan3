# 1. Undirected / Unweighted

```python
import pandas as pd
```

`# Ex` – default

```python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
g = torch_geometric.data.Data(
    edge_index = links
)
```

```python
graft.graph.plot_undirected_unweighted(g)
```

`# Ex` – node_names

```python
graft.graph.plot_undirected_unweighted(
    g, 
    node_names = ['a','b','c','d','e'], 
)
```

`# Ex` – node_color (continuous)

```python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
g = torch_geometric.data.Data(
    edge_index = links, 
    y = np.random.randn(5)
)
```

```python
graft.graph.plot_undirected_unweighted(
    g,
    node_color=g.y
)
```

`# Ex` – node_color (discrete)

```python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
g = torch_geometric.data.Data(
    edge_index = links, 
    y = torch.tensor([0,1,0,0,1])
)
```

```python
graft.graph.plot_undirected_unweighted(
    g,
    node_color=g.y
)
```

`# Ex` – node_color (discrete) / node_size

```python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
g = torch_geometric.data.Data(
    edge_index = links, 
    y = torch.tensor([0,1,0,0,1]),
    x = torch.tensor([10,100,15,20,150])
)
```

```python
graft.graph.plot_undirected_unweighted(
    g,
    node_color=g.y,
    node_size=g.x
)
```

`# Ex` – draw options

```python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
g = torch_geometric.data.Data(
    edge_index = links
)
```

```python
graft.graph.plot_undirected_unweighted(
    g,
)
```

```python
dr_opts = {
    'vertex_size':30
}
graft.graph.plot_undirected_unweighted(
    g,
    draw_options= dr_opts
)
```

```python
dr_opts = {
    'edge_marker_size': 200, 
    'output_size': (300,300)
}
graft.graph.plot_undirected_unweighted(
    g,
    draw_options= dr_opts
)
```

# 2. Undirected / Weighted

`# Ex1` – 무방향 그래프, 가중치 값이 존재

```python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
weights = torch.tensor([5, 5, 1.5, 1.5, 0.19, 0.19], dtype=torch.float)

g = torch_geometric.data.Data(
    edge_index=links,
    edge_attr=weights,
)
```

```python
graft.graph.plot_undirected_weighted(
    g,
)
```

`# Ex2` – 무방향 그래프, 가중치 값이 존재

```python
links = torch.tensor([[0, 1, 2, 3, 4, 3],
                      [1, 0, 3, 2, 3, 4]], dtype=torch.long)
weights = torch.tensor([5, 5, 1.5, 1.5, 0.19, 0.19], dtype=torch.float)

graph = torch_geometric.data.Data(
    edge_index=links,
    edge_attr=weights,
    y = torch.tensor([1,1,0,0,0])
)
```

```python
graft.graph.plot_undirected_weighted(graph,node_color=graph.y)
```
