```python
import numpy as np
import matplotlib.pyplot as plt
```

`-` 아이디어: 임의의 real-valued signal은 짝함수와 홀함수로 나눌 수 있다. 

$$f=f_{e}+f_{o}$$

`-` 짝함수를 퓨리에변환하면 언제나 실수&EVEN이 나온다. 

(예시1)

```python
t=np.linspace(-3*np.pi,3*np.pi,100)
f=np.cos(t)
```

```python
plt.plot(f)
```

```python
fhat = np.fft.fft(f)
plt.plot(abs(fhat))
```
