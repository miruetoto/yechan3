```python
import pandas as pd
import numpy as np
import torch
from sklearn.preprocessing import RobustScaler
```

```python
df_train = pd.read_csv("https://raw.githubusercontent.com/guebin/DL2022/main/posts/IV.%20RNN/2022-12-07-DailyDelhiClimateTrain.csv").drop('date',axis=1)
df_test = pd.read_csv("https://raw.githubusercontent.com/guebin/DL2022/main/posts/IV.%20RNN/2022-12-07-DailyDelhiClimateTest.csv").drop('date',axis=1)
```

```python
for colname in df_train:
    df_train[[colname]] = RobustScaler().fit(df_train[[colname]]).transform(df_train[[colname]])
for colname in df_test:
    df_test[[colname]] = RobustScaler().fit(df_test[[colname]]).transform(df_test[[colname]])    
```

```python
df_train
```

```python
df_test
```

```python
class ClimateDatasets(torch.utils.data.Dataset):
    def __init__(self,train=False,validate=False, test=False, window_size=7):
        ## 1. load the data 
        self.df_train = df_train 
        self.df_test = df_test 
        ## 2. define x and y 
        if train: 
            X = self.df_train 
            y = self.df_train.meantemp 
        else: 
            X = self.df_test
            y = self.df_test.meantemp

        self.x, self.y = [], []  
        for i in range(len(X) - window_size):
            v = X.iloc[i:(i+window_size)].values
            self.x.append(v)
            self.y.append(y.iloc[i+ window_size]) 
    def __len__(self):
        return self.num_sample 
    def __getitem__(self,index):
        x = self.x[index].astype(np.float32)
        y = self.y[index].astype(np.float32)
        return x,y 
```

```python
ds = ClimateDatasets(train=True)
```
